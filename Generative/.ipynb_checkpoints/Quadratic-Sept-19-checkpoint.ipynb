```python
import torch
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
from torchvision.utils import make_grid
import torch.nn.functional as F

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#device = 'cpu'
print(device)
```

```python
def load_dataset():
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
    #dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)

    return dataset
```

```python
def control_estimate(x, y_samples, t, beta):

    batch_size, channels, height, width = x.shape
    num_samples = y_samples.shape[0]
    
    x_flat = x.view(batch_size, -1)
    y_flat = y_samples.view(num_samples, -1)

    # Compute hyperbolic functions
    sqrt_beta = torch.sqrt(torch.tensor(beta, device=device))
    cosh_term = torch.cosh((1 - t) * sqrt_beta)
    sinh_term = torch.sinh((1 - t) * sqrt_beta)
    coth_term = cosh_term / sinh_term
    
    x_squared = (x_flat ** 2).sum(dim=1, keepdim=True)
    y_squared = (y_flat ** 2).sum(dim=1, keepdim=True)
    xy_term = torch.matmul(x_flat, y_flat.t())

    log_exponent = -sqrt_beta * ((x_squared + y_squared.t()) * coth_term * 0.5 - xy_term / sinh_term )
    log_exponent += - y_squared.t() * sqrt_beta * torch.cosh(sqrt_beta) / (2 * torch.sinh(sqrt_beta))
    weights = F.softmax(log_exponent, dim=0)  # Shape: (num_samples, batch_size)
    

    # Compute x_hat using convex combination
    #x_hat = (weights[:, :, None] * y_flat[:, None, :]).sum(dim=0)
    x_hat = torch.einsum('ns,nd->sd', weights.transpose(0, 1), y_flat)
    
    # Compute the score
    u = (sqrt_beta / sinh_term) * (x_hat - x_flat * cosh_term)
    
    return u.view(batch_size, channels, height, width), x_hat.view(batch_size, channels, height, width)
```

```python
 #Main execution
dataset = load_dataset()
y_samples = torch.stack([img for img, _ in dataset]).to(device)
```

```python
beta = 1
eps = 1e-3
num_steps = 512
_, chan, w, h = y_samples.shape
batch_size = 2
t = torch.linspace(0, 1-eps, num_steps, device=device)
delta_t = t[1] - t[0]
x = torch.zeros(batch_size, chan, w, h, device= device)
s = torch.zeros(num_steps, batch_size, chan, w, h, device= device)
ss = torch.zeros(num_steps, batch_size, chan, w, h, device= device)

for  i, tt in enumerate(t):
    u, cvx_comb = control_estimate(x, y_samples, tt, beta)
    x  = x + u * delta_t + torch.randn_like(x, device=device) * torch.sqrt(delta_t)
    s[i, :, :, :, :] = cvx_comb
    ss[i, :, :, :, :] = x
```

```python
def plot_images(images):
    images =  (images * 127.5 + 128).clip(0, 255).to(torch.uint8)
    sample_grid = make_grid(images, nrow=int(np.sqrt(images.shape[0])))
    plt.figure(figsize=(2,2))
    plt.axis('off')
    plt.imshow(sample_grid.cpu().permute(1, 2, 0).squeeze())
    plt.show()

plot_images(x)
```
